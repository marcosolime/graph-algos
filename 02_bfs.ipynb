Breath-first search
- it is implemented with a queue
- a queue is a LIFO data structure â€” like a line of people; we add from right, we pop from left
- queues are implemented in the `deque` module

```python
from collections import deque

# graph definition
graph = {
    'a': ['c', 'b'],
    'b': ['d'],
    'c': ['e'],
    'd': ['f'],
    'e': [],
    'f': []
}
```

```python
def bfs(graph, src):
    queue = deque()
    queue.append(src)

    while len(queue) > 0:
        cur = queue.popleft()
        print(cur)
        neighbors = graph[cur]
        for neigh in neighbors:
            queue.append(neigh)

bfs(graph, 'a')
```

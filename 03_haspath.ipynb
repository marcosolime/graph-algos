Has Path
- we check if there exists a path between two nodes in a graph
- we return `True` if exists, `False` otherwise

```python
# graph definition
graph = {
    'a': ['c', 'b'],
    'b': ['d'],
    'c': ['e'],
    'd': ['f'],
    'e': [],
    'f': []
}
```

Recursive approach

```python
def has_path(graph, src, dest):
    if src == dest:
        return True
    
    for neigh in graph[src]:
        if has_path(graph, neigh, dest):
            return True
    
    return False

print(has_path(graph, 'a', 'e'))
print(has_path(graph, 'e', 'a'))
```

Depth-first search

```python
# graph definition
graph = {
    'a': ['c', 'b'],
    'b': ['d'],
    'c': ['e'],
    'd': ['f'],
    'e': [],
    'f': []
}
```

Recursive approach

```python
def dfs(src, graph):
    print(src)

    neighbors = graph[src]
    for neigh in neighbors:
        dfs(neigh, graph)

dfs('a', graph)
```

Iterative approach *(stack)*
- a Python list has `.append()` and `.pop()` methods
- `.append()` adds an element at the top of the stack
- `.pop()` removes and returns the element at the top of the stack

```python
def dfs(src, graph):
    stack = [src]

    while len(stack) > 0:
        cur = stack.pop()
        print(cur)
        
        neighbors = graph[cur]
        for neigh in neighbors:
            stack.append(neigh)

dfs('a', graph)
```
